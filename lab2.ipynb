## Group and Visulize the Data

To group and visualize the data we will take two columns and use group by to generate two statistics. Then we will make them into visualizations using matplot.

```python
summary = df.groupby('Year')['Deaths'].agg(
    Total_Deaths='sum',
    Average_Deaths='mean'
).reset_index()

print(summary)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(summary['Year'], summary['Total_Deaths'])
plt.title('Total Deaths by Year')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.show()
```

This data shows a trend of increasing deaths per average by year. The peak is in 2017 with 347173 deaths and the low was 82470 in 1999.

```python
plt.figure(figsize=(6,4))
plt.scatter(summary['Year'], summary['Average_Deaths'])
plt.title('Average Deaths by Region')
plt.xlabel('Year')
plt.ylabel('Average Deaths')
plt.show()
```

This scatterplot tells a similar story but uses the average deaths instead. As we can see, there is a steady increase in deaths by year.
